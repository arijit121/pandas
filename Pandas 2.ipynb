# Q.8. Write a function that returns a team v/s team record.

```python
match
```

```python
def func(t1,t2):
    mask1 = match['team1'] == t1
    mask2 = match['team2'] == t2
    mask3 = match['team1'] == t2
    mask4 = match['team2'] == t1

    print(match[(mask1 & mask2) | (mask3 & mask4)]['winner'].value_counts())
    
func("Kolkata Knight Riders", "Royal Challengers Bangalore")
```

## Q.6 Find the Orange Cap holder ?

```python
# We see season is in match dataframe. So we have to join match and delivery dataframe. Just like JOIN in SQL, we need a common 
# column. Here it's match_id and id in delivery and match respectivley

# left_on : common cloumn name in old dataframe
# right_on : common column name in to-be-merged dataframe

df = match.merge(delivery,left_on = 'id', right_on = 'match_id')
df
```

```python
# We need to find how many runs per batsman per season.
# multiple groupby

df.groupby(['season','batsman']).sum()['batsman_runs'].sort_values(ascending = False).reset_index().drop_duplicates(subset=['season'], keep = 'first')
```

<h1 style = "color:#FF0000">Q. Which batsman has highest Strike Rate in last 5 overs(16-20) in each IPL season ?</h1>

```python
delivery
```

```python
mask1 = delivery['over'] >= 16
mask2 = delivery['over'] <= 20

df1 = delivery[mask1 & mask2]
df1
```

```python
new = df1['batsman'].value_counts()
```

```python
batsman_list = new[new>200].index.tolist()
```

```python
final = df1[df1['batsman'].isin(batsman_list)]
final
```

```python
runs = final.groupby('batsman').sum()['batsman_runs']
runs
```

```python
balls = final['batsman'].value_counts()
balls
```

```python
s_rate = (runs/balls)*100
s_rate.sort_values(ascending = False).plot(kind = 'bar')
s_rate.sort_values(ascending = False)[[0]]
```
