```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
# 1.pie chart
# 2.line chart
# 3.scatter plot
# 4.bar chart
# 5.histogram
```

## 1.pie charts

```python
x=delivery[delivery['batsman']=='V Kohli']
```

```python
y=x['batsman_runs'].value_counts()
```

```python
plt.pie(y,labels=y.index,autopct="%0.2f%%")
plt.show()
```

## 2.line chart

```python
big=match.merge(delivery,left_on='id',right_on='match_id')
```

```python
batsman_data=big[big['batsman']=='V Kohli']
batsman_data1=big[big['batsman']=='RG Sharma']
```

```python
performance=batsman_data.groupby('season').sum()['batsman_runs']
performance1=batsman_data1.groupby('season').sum()['batsman_runs']
```

```python
performance
plt.plot(performance.index,performance.values,color="red",marker='D',label="V kohli")
plt.plot(performance1.index,performance1.values,color="green",marker='D',label="Rg sharma")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.title("Batsman Season by season Performance")
plt.legend()
```

```python
def compatator(batsman1,batsman2):
    big=match.merge(delivery,left_on='id',right_on='match_id')
    batsman_data=big[big['batsman']==batsman1]
    batsman_data1=big[big['batsman']==batsman2]
    performance=batsman_data.groupby('season').sum()['batsman_runs']
    performance1=batsman_data1.groupby('season').sum()['batsman_runs']
    plt.plot(performance.index,performance.values,color="red",marker='D',label="batsman1")
    plt.plot(performance1.index,performance1.values,color="green",marker='D',label="batsman2")
    plt.xlabel('Season')
    plt.ylabel('Runs Scored')
    plt.title("Batsman Season by season Performance")
    plt.legend()
```

```python
compatator("MS Dhoni","V Kohli")
```

```python
plt.scatter(performance.index,performance.values)
```
